```python
import collections
from transformers import AutoTokenizer
import pandas as pd

# Function to count occurrences of words using built-in libraries
def count_words(text):
    words = text.lower().split()  # Convert to lowercase and split into words
    word_count = collections.Counter(words)
    return word_count

# Function to count unique tokens using Transformers' Auto Tokenizer
def count_unique_tokens(text, model_name):
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    tokens = tokenizer.tokenize(tokenizer.encode(text, add_special_tokens=False))
    token_count = collections.Counter(tokens)
    return token_count

# Read text from file
with open('/content/drive/MyDrive/assignment/output_text (1).txt', 'r', encoding='utf-8') as file:
    text = file.read()

# Task 3.1
word_count = count_words(text)
top_30_words = word_count.most_common(30)

# Store Top 30 words and their counts into a CSV file
df_words = pd.DataFrame(top_30_words, columns=['Word', 'Count'])
df_words.to_csv('top_30_words.csv', index=False)

# Task 3.2
model_name = 'textattack/bert-base-uncased-MNLI'  # Replace with your desired model
token_count = count_unique_tokens(text, model_name)
top_30_tokens = token_count.most_common(30)

# Print or further process the Top 30 tokens as needed
print("Top 30 Tokens:", top_30_tokens)
```
